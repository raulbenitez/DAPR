```python
import seaborn as sns

df = sns.load_dataset("car_crashes")
df1 = df.dropna() # remove cars with missing values
df2  = df1.select_dtypes(include='number') # remove non-numerical colunms

df2.head()
```

```python
df2.shape
```

```python
g = sns.pairplot(df2)
```

```python
from sklearn.cluster import KMeans
group = KMeans(n_clusters=2).fit_predict(df2)

df3 = df2.copy()
df3['kmeans'] = group
g = sns.pairplot(df3, hue="kmeans")
```

```python
from sklearn.cluster import AgglomerativeClustering
group = AgglomerativeClustering(n_clusters=3).fit_predict(df2)

df3 = df2.copy()
df3['AG'] = group
g = sns.pairplot(df3, hue="AG")
```
